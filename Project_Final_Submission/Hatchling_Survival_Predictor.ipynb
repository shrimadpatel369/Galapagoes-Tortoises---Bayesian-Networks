```python

import sys
import pandas as pd
from PyQt5.QtWidgets import QApplication, QGraphicsScene, QMainWindow,QDialog,QLabel,QVBoxLayout,QMessageBox
from PyQt5.QtGui import QImage, QPixmap
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure
from PyQt5.uic import loadUi
import QueryRunner as qr


class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        loadUi('Predictor.ui', self)  # Replace with the path to your .ui file
        
        egg_df = pd.read_csv('../Final_Data/Egg_Condition_Final.csv')
        hatchling_df = pd.read_csv('../Final_Data/Hatchling_Data_final.csv')
        female_df = pd.read_csv('../Final_Data/Female_Condition_Final.csv')


        self.egg_model = qr.EggModel(egg_df)
        self.hatchling_model = qr.HachlingModel(hatchling_df)
        self.female_model = qr.FemaleModel(female_df)
        self.initializeNesting()

        self.network_hatchling.clicked.connect(lambda: self.showNetwork('hatchling'))
        self.network_egg.clicked.connect(lambda: self.showNetwork('egg'))
        self.network_female.clicked.connect(lambda: self.showNetwork('female'))
        
        self.process_hatchling.clicked.connect(self.solveHatchling)
        self.process_egg.clicked.connect(self.solveEgg)
        self.process_female.clicked.connect(self.solveFemale)
        

        self.clear_egg.clicked.connect(lambda: self.clear('egg'))
        self.clear_hatchling.clicked.connect(lambda: self.clear('hatchling'))
        self.clear_female.clicked.connect(lambda: self.clear('female'))
        

    def initializeNesting(self):
        self.nesting_hatchling.addItems(self.hatchling_model.category_codes.keys())

    def checkEgg(self):
        return self.ndvi_egg.text()!='' and self.temp_egg.text()!=''
    
    def checkHatchling(self):
        return self.ndvi_hatchling.text()!='' and self.initialLength_hatchling.text()!='' and self.initialMass_hatchling.text()!='' and self.length_hatchling.text()!='' and self.weight_hatchling.text()!='' and self.nesting_hatchling.currentText()!=''
    def checkFemale(self):
        return self.elevation_female.text()!='' and self.bodyCondition_female.text()!=''
    def validateEgg(self):
        try:
            float(self.ndvi_egg.text())
            float(self.temp_egg.text())
            return True
        except:
            return False
    
    def validateHatchling(self):
        try:
            float(self.ndvi_hatchling.text())
            float(self.initialLength_hatchling.text())
            float(self.initialMass_hatchling.text())
            float(self.length_hatchling.text())
            float(self.weight_hatchling.text())
           
            return True and (self.nesting_hatchling.currentText() in self.hatchling_model.category_codes.keys())
        except:
            return False

    def validateFemale(self):
        try:
            float(self.elevation_female.text())
            float(self.bodyCondition_female.text())
            return True
        except:
            return False
        


    
    def show_warning(self, title, message):
        # Create a QMessageBox for warning
        msg_box = QMessageBox()
        msg_box.setIcon(QMessageBox.Warning)
        msg_box.setWindowTitle(title)
        msg_box.setText(message)
        msg_box.setStandardButtons(QMessageBox.Ok)
        msg_box.exec_()  # Show the message box

    def solveHatchling(self):
        ndvi = self.ndvi_hatchling.text()
        initialMass_hatchling = self.initialMass_hatchling.text()
        weight_hatchling = self.weight_hatchling.text()
        length_hatchling = self.length_hatchling.text()
        initialLength_hatchling = self.initialLength_hatchling.text()  
        nesting_hatchling = self.nesting_hatchling.currentText()

        if self.validateHatchling():
            
            if self.checkHatchling():
                result = self.hatchling_model.process(ndvi=ndvi,nesting = nesting_hatchling,weight=weight_hatchling,length=length_hatchling,initialLength=initialLength_hatchling,initialMass=initialMass_hatchling)

                df = pd.DataFrame(result.values,columns=['x'])

                self.show_graph(result,'hatchling')
            else:
                self.show_warning("Missing Arguments", "Please enter all required values.")
                print("Please fill all the required fields")
        else:
            self.show_warning("Invalid Arguments", "NDVI -> float\nInitial Mass -> float\n Length -> float \n Initial Length -> float\n Weight -> float\n Nesting -> select from provided values only")

    def solveEgg(self):
        ndvi = self.ndvi_egg.text()
        temp = self.temp_egg.text()
        if self.validateEgg():
            if self.checkEgg():
                result = self.egg_model.process(ndvi=ndvi, temp=temp)
                df = pd.DataFrame(result.values,columns=['x'])

                self.show_graph(df,'egg')
            else:
                self.show_warning("Missing Arguments", "Please enter all required values.")
                print("Please fill all the required fields")
        else:
            self.show_warning("Invalid Arguments", "NDVI -> float\nTemperature -> float")

    def solveFemale(self):
        bci = self.bodyCondition_female.text()
        elevation = self.elevation_female.text()

        if self.validateFemale():
            if self.checkFemale():
                result = self.female_model.process(bci=bci, elevation=elevation)
                df = pd.DataFrame(result.values, columns=['x'])
                self.show_graph(df,'female')
            else:
                self.show_warning("Missing Arguments", "Please enter all required values.")
                print("Please fill all the required fields")
        else:
            self.show_warning("Invalid Arguments", "BCI -> float\nElevation -> float")
    
    def showNetwork(self,name):
        # Create a QDialog to display the image
        dialog = QDialog(self)
        dialog.setWindowTitle("Image Display")

        # Create a QLabel and load the image
        image_label = QLabel(dialog)
        if name=='egg':

            pixmap = QPixmap("../Networks/Egg_Model.png")  # Replace with the actual path to your image
        elif name=='hatchling':
            pixmap = QPixmap("../Networks/Hatchling_Model.png")  # Replace with the actual path to your image
        else:
            pixmap = QPixmap("../Networks/Female_Model.png")  # Replace with the actual path to your image

        image_label.setPixmap(pixmap)
        # Optionally, set QLabel properties for better scaling
        image_label.setScaledContents(True)
        image_label.setFixedSize(pixmap.size().width(), pixmap.size().height())  # Set this to the desired size of the image in the dialog

        # Arrange QLabel in a QVBoxLayout and set the layout in the dialog
        layout = QVBoxLayout()
        layout.addWidget(image_label)
        dialog.setLayout(layout)

        # Show the dialog
        dialog.exec_()
    def clear(self,name):
        scene = QGraphicsScene()

        if name=='egg':
            self.ndvi_egg.clear()
            self.temp_egg.clear()
            self.result_Egg.setScene(scene)
        elif name=='hatchling':
            self.ndvi_hatchling.clear()
            self.initialLength_hatchling.clear()
            self.initialMass_hatchling.clear()
            self.length_hatchling.clear()
            self.weight_hatchling.clear()
            self.result_Hatchling.setScene(scene)
        else:
            self.elevation_female.clear()
            self.bodyCondition_female.clear()
            self.result_female.setScene(scene)
            

    def show_graph(self, df, name):
    # Create matplotlib figure and plot
        fig = Figure(figsize=(8, 5))
        canvas = FigureCanvas(fig)
        ax = fig.add_subplot(111)

        # Choose parameters based on the 'name' argument
        if name == 'hatchling':
            days_alive = df.index
            phi_days_alive = df['x']
            bars = ax.bar(days_alive, phi_days_alive, color='skyblue')
            ax.set_xlabel('Days Alive')
            ax.set_ylabel('Probability (phi(days_alive))')
            ax.set_title('Probability Distribution of phi(days_alive) over Days')
        elif name == 'egg':
            days_alive = df.index
            phi_days_alive = df['x']
            bars = ax.bar(days_alive, phi_days_alive, color='green')
            ax.set_xlabel('Egg Survival')
            ax.set_ylabel('Probability (phi(Egg_survival))')
            ax.set_title('Probability Distribution of phi(Egg_survival) over Days')
        else:
            days_alive = df.index
            phi_days_alive = df['x']
            bars = ax.bar(days_alive, phi_days_alive, color='red')
            ax.set_xlabel('Eggs Detected')
            ax.set_ylabel('Probability (phi(egg_detected))')
            ax.set_title('Probability Distribution of phi(egg_detected) over Days')
        
        # Set x-ticks
        ax.set_xticks(days_alive)

        # Add annotations for each bar
        for bar in bars:
            yval = bar.get_height()  # Get the height of the bar (i.e., the value of phi)
            ax.annotate(
                f'{yval:.4f}',  # Format the label to 4 decimal places
                xy=(bar.get_x() + bar.get_width() / 2, yval),  # Position at the center top of each bar
                xytext=(0, 5),  # Offset the text slightly above the bar
                textcoords="offset points",
                ha='center',
                va='bottom'
            )

        # Render to QImage
        canvas.draw()
        width, height = fig.get_size_inches() * fig.get_dpi()
        qimage = QImage(canvas.tostring_rgb(), int(width), int(height), QImage.Format_RGB888)
        
        # Display QImage in QGraphicsView based on the 'name'
        scene = QGraphicsScene()
        scene.addPixmap(QPixmap.fromImage(qimage))
        if name == 'egg':
            self.result_Egg.setScene(scene) 
        elif name == 'hatchling':
            self.result_Hatchling.setScene(scene)
        else:
            self.result_female.setScene(scene)


app = QApplication(sys.argv)
window = MainWindow()
window.show()
sys.exit(app.exec_())
```
